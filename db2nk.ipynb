```python
for c in nc:
    Q1=df1[c].quantile(0.25)
    Q3=df1[c].quantile(0.75)
    IQR=Q3-Q1
    lowerbound=Q1-1.5*IQR
    upperbound=Q3+1.5*IQR
    df1[c]=np.where((df1[c]<lowerbound)|(df1[c]>upperbound),df1[c].median(),df1[c])
df1
```
