```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
df.hist(figsize=(10,6),edgecolor='black',grid=False)
plt.suptitle('Histograms of iris dataset features',fontsize=16)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))
for i,column in enumerate(df.columns[:-1],1):    #skipping species column
    plt.subplot(2,2,i)
    sns.boxplot(y=df[column],color='skyblue')
    plt.title(f"Boxplot of {column}")
plt.tight_layout()
plt.show()
```

```python
sns.histplot(x='species',data=df)
```

```python
sns.histplot(x='petal_length',hue='species',data=df)
```

```python
sns.histplot(x='petal_width',hue='species',kde=True,multiple='dodge',data=df)
```
