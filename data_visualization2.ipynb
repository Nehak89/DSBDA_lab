```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
df=pd.read_csv("/home/student/Documents/Titanic-Dataset.csv")
```

```python
df.head()
```

```python
#Were younger passengers more likely to survive?
ax=df.boxplot(column='Age',by='Survived',figsize=(10,5))
ax.set_ylabel('age')
ax.set_title("Boxplot:Survived VS Age")
```

```python
#Were younger passengers more likely to survive considering age?
ax=df.boxplot(column='Age',by=['Survived','Sex'],figsize=(10,5))
ax.set_ylabel('age')
ax.set_title("Boxplot:Survived VS Age")
```

```python
#Did survival rates differ by boarding location and age?
ax=df.boxplot(column='Age',by=['Embarked','Survived'],figsize=(10,5))
ax.set_ylabel('age')
ax.set_title("Boxplot:Embarked VS Survived")
```

```python
#did first-class passengers have a higher survival rate?
ax=df.boxplot(column='Pclass',by='Survived',figsize=(10,5))
ax.set_ylabel('Pclass')
ax.set_title("Boxplot:Pclass VS Suvived")
```

```python
#passengers who paid high survived?
ax=df.boxplot(column='Fare',by='Survived',figsize=(10,5))
ax.set_ylabel('fare')
ax.set_title("Boxplot:fare VS Suvived")
```
